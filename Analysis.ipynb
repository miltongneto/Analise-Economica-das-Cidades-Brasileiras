# Analise Econômica das Cidades Brasileiras

### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

### Load dataset

```python
data = pd.read_csv('resources/BRAZIL_CITIES.csv', sep=';')
print('Shape:', data.shape)
data.head()
```

### Change the types (clean and transform)

```python
data.select_dtypes('O')
```

```python
cols_to_transform = data.select_dtypes('O').drop(['CITY', 'STATE', 'RURAL_URBAN', 'GVA_MAIN', 'REGIAO_TUR', 'CATEGORIA_TUR'], axis=1).columns
data.loc[:, cols_to_transform].head()
```

```python
data.loc[:, cols_to_transform] = data.loc[:, cols_to_transform].apply(lambda x: x.str.replace(',','.').astype(float))
```

### Missing Values

```python
missing_values = data.isnull().sum()
```

#### Percentage of columns with missing data

```python
missing_values[missing_values > 0].size / data.shape[1] * 100
```

#### Distribution of missing values in columns

```python
missing_values.hist(bins=25, figsize=(8,6))
```

#### Columns without missing values

```python
missing_values[missing_values == 0]
```

## Analysis of States  

```python
state_group = data.groupby('STATE')
```

#### Number of cities per state

```python
state_group.size()
```

```python
state_group.size().hist(figsize=(8,6))
```

#### Population of the states
There are two variables that represent the Population. So, we will analyze and compare both.

```python
population_df = pd.DataFrame([state_group['ESTIMATED_POP'].sum(), state_group['POP_GDP'].sum()]).T
```

```python
fig, ax = plt.subplots(sharey=True, figsize=(12, 8))

sns.lineplot(data=population_df, y='ESTIMATED_POP', x=population_df.index, ax=ax, label='ESTIMATED_POP')
sns.lineplot(data=population_df, y='POP_GDP', x=population_df.index, ax=ax, label='POP_GDP')
```

```python
population_df.corr()
```

#### Very high correlation
Now, choose one and analyze the distribution.

```python
population_df['POP_GDP'].hist(bins=15, figsize=(12,8))
```

```python
population_df['POP_GDP'].sort_values().plot(kind='barh', figsize=(12,10))
```

### Gross Domestic Product

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))
state_group['GDP'].sum().hist(bins=15, ax=ax[0])
state_group['GDP_CAPITA'].sum().hist(bins=15, ax=ax[1])
ax[0].title.set_text('Histogram of GDP (SUM)')
ax[1].title.set_text('Histogram of GDP_CAPITA (SUM)')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))
state_group['GDP'].sum().plot(kind='barh', ax=ax[0])
state_group['GDP_CAPITA'].sum().plot(kind='barh', ax=ax[1])
ax[0].title.set_text('GDP TOTAL PER STATE (SUM)')
ax[1].title.set_text('GDP_CAPITA TOTAL PER STATE (SUM)')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))
state_group['GDP'].mean().hist(bins=15, ax=ax[0])
state_group['GDP_CAPITA'].mean().hist(bins=15, ax=ax[1])
ax[0].title.set_text('Histogram of GDP (MEAN)')
ax[1].title.set_text('Histogram of GDP_CAPITA (MEAN)')
```

```python
state_group['GDP_CAPITA'].mean().sort_values().plot(kind='barh', figsize=(12,10), title='Average Gross Domestic Product per capita')
```

### Relation between Gross Domestic Product and Population 

```python
gdp_pop_df = pd.concat([population_df['POP_GDP'], state_group['GDP'].sum()], axis=1)

fig, ax = plt.subplots(sharey=True, figsize=(12, 8))
sns.regplot(data=gdp_pop_df, x='POP_GDP', y='GDP', ax=ax)
```

```python
gdp_pop_df.corr('spearman') # Evaluate the intensity of the correlation (rank)
```
