### Normality test

```python
def compute_normality_test(sample):
    mean = sample.mean()
    std = sample.std()
    n = sample.count()
    
    ks = stats.kstest(sample.values, 'norm', N=n, args=(mean, std))
    shapiro = stats.shapiro(sample.values)
    return (ks, ('ShapiroTestResult', shapiro))
```

```python
print('PE')
print(compute_normality_test(pe_df['GDP_CAPITA'].sample(50).dropna()))

print('PB')
print(compute_normality_test(pb_df['GDP_CAPITA'].sample(50).dropna()))

print('SE')
print(compute_normality_test(se_df['GDP_CAPITA'].sample(50).dropna()))

print('RN')
print(compute_normality_test(rn_df['GDP_CAPITA'].sample(50).dropna()))
```

In all tests (KS and Shapiro-Wilk for PE, PB, SE and RN), the p-value was less than 0.05 (significance). Then, the null hypothesis is rejected and there is evidence that the data <b>is not from a normally distributed population

#### IDHM_Educacao

```python
sample_IDHM_Educacao = data['IDHM_Educacao'].dropna().sample(500)
sample_IDHM_Educacao.hist(bins=20)

print('IDHM_Educacao')
print(compute_normality_test(sample_IDHM_Educacao))
```

In KS and Shapiro-Wilk tests, the results were p-value > 0.05. We cannot reject the null hypothesis, then there isn't strong evidence that the distribution is different <b>from the normal distribution.

### Hypothesis tests

#### Comparing the GDP_CAPITA
The samples aren't from a normal distribution, then we let's use the nonparametric statistical hypothesis tests, the <b> Mann–Whitney U test

H0: PE = PB 

H1: PE ≠ PB

```python
stats.mannwhitneyu(pe_df['GDP_CAPITA'].dropna().values, pb_df['GDP_CAPITA'].dropna().values)
```

Don't reject the null hypothesis

H0: SE = RN

H1: SE > RN

```python
stats.mannwhitneyu(se_df['GDP_CAPITA'].dropna().values, rn_df['GDP_CAPITA'].dropna().values, alternative='greater')
```

Reject the null hypothesis

#### Comparing the IDHM_Educacao
The sample is from a normal distribution, then we let's use the parametric statistical hypothesis tests, z-Test, because the sample size is big.

IDHM_Educacao

H0: BR = PE_MEAN

```python
idhm_educacao_pe_mean = data.groupby('STATE')['IDHM_Educacao'].mean()['PE']
```

```python
stests.ztest(sample_IDHM_Educacao.values, value=idhm_educacao_pe_mean)
```
